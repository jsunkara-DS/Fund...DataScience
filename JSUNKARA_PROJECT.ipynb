# DESCRIPTION

The dataset is from https://www.kaggle.com/gregorut/videogamesales  
The dataset contains the sales information of different vediogames globally and the attributes are as follows  
Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

# TRAINING MODEL

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import LabelEncoder

data1 = data.copy()

le = LabelEncoder()

feature = ["Platform", "Genre"]


for col in feature:
    data1[col] = le.fit_transform(data[col])
```

```python
data1.head()
```

```python
X = data1[['Platform', 'Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].values

y = data1['Global_Sales'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import RobustScaler
ro = RobustScaler()
x_train = ro.fit_transform(x_train)
x_test = ro.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score,classification_report, roc_curve
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
```

```python
regressor.score(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
yTreePre= regressor.predict(x_test)
```

```python
r2Tree= r2_score(y_test ,yTreePre )
r2Tree
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(yTreePre, y_test)
print('Mean absolute error '+str(mae))
```

```python
db = pd.DataFrame({'Actual': y_test, 'Predicted': yTreePre})
db
```

```python
db1 = db.head()
db1.plot(kind='line',figsize=(15,8))
plot.show()
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
yLinearPre = lr.predict(x_test)
```

```python
r2Linear = r2_score(y_test , yLinearPre)
r2Linear
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(yLinearPre, y_test)
print('Mean absolute error '+str(mae))
```

```python
dbLinear = pd.DataFrame({'Actual': y_test, 'Predicted': yLinearPre})
dbLinear
```

```python
dbLinear1 = dbLinear.head()
dbLinear1.plot(kind='line',figsize=(15,8))
plot.show()
```
